### Vim

Vim is a modal editor, this means that it has different kinds of modes it operates in. This is the basic idea behind vim, its that when you are using a text editor, you are doing different things like inserting or searching.

When we start vim, it starts in something called normal mode and all the key combinations work in a certain way. Then there are keys that help you switch to another mode where key combinations have different meanings.


Noraml to insert -> i
insert to normal -> esc

normal mode is for navigating things, going from file to file and so on and in the insert mode you type in text.

Other modes:
replace mode - instead of insert you replace things
normal to replace -> r


some modes for selection
visual which further has visual line and visual block
normal to visual -> v
normal to line visual -> shift+v
normal to block visual -> cntrl+v

command line mode -> : (colon key)

**todo**: configure vim to show line numbers.

The bottom left of vim always shows you what mode you are in unless you are in normal mode.

Vim believes that using the mouse is inefficient hence all the functionality is built in. Hence we can go to command mode using the : colon key. Now this is vim shell not the regular shell that we use hence we have to use vim commands.

Short example

```bash
>vim test.md
```

open the test.md file

```bash
>i
```
switch to insert mode and enter a bunch of text


```bash
>Esc
```
switch back to normal mode

```bash
>:quit
```
go to command mode and quit the program (instead of using mouse, going to some menu and clicking quit)

Instead of using :quit we can just do :q and that'll do the same thing.

How do we save the file?
Well we can go to insert mode, insert a bunch of text, press esc to go back to normal mode and then do
:w, where w stands for write. We can then do :q to quit and then reopen the file and check its contents

to look up the documentation we can use the :help command

so we can do

```bash
:help :w 
```

to get the documentation of what :w does. We can then do :q and it will close that window

you can make small point edits in normal mode

u is used for undo. 

gg is to go to the top of the file

If we use a visual editor like sublime or vs code we can open multiple files and tabs
Vim maintains a set of open buffers, that's the word it uses for open files

For example vim has windows where you can look at different parts of the same file, add some imports while
you write some code below

Vim has a number of tabs and each tab has some number of windows and each window corresponds to some buffer. A particular buffer can be open in 0 or more buffers at a time.

:q is actually quit the current window and when there are no more windows to close it will close the tab.

Vim's interface is a programming language, different key combinations have different meanings and once you learn them, you can glue them together to do different things just like you would in a programming language

j - down
k - up
h - left
l - right

w - move by one word
b - backward by one word
e - end of a word

0 - beginning of line
$ - end of line
^ - first non empty character of the line

Cntrl + U - scroll up
Cntrl + D - scroll down

G - all the way down
gg - all the way up

l - lowest line on your screen
m - middle
h - highest

if I am on some line and I want to find the first o, I can press fo and it will search the first o
similary Fo will find backward

there's a variant to this, to will jump to one letter before the first o and To will do the same backward

in normal mode:
o - creates a new line below where you are and puts you into insert mode, similarly there's O to insert line above and put you in insert mode.

To delete words we use the d command. However, d does not work by itself you have to give it some movement commands (h,j,k,l,w,b and so on)

if you are in the middle of the word and you want to delete till the end you can do de

Similary you can use c instead of d for change, if you do ce it'll delete till the end of the word but also put you in insert mode for you to insert something

dd - delete the line
cc - deletes the line, but leaves it blank and goes to insert mode

in normal mode
x - delete that character

if we are on a character and we press r, it takes another character and replaces the one we are on with the new one

for eg > writing and the cursor is on r, if we press ra it will become waiting

u for undo
Cntrl + R = redo

y for copying and p for pasting

y also takes direction like delete

yw copy the word

yy copy the current line 

you can enter visual mode by pressing v
then you can do the normal key movements and it will select all that text

then you can do y for copy and it will take you back to normal mode

then you can go somewhere else and do p for paste

similarly there's visual line mode that selects lines at a time - V
and visual block - Cntrl + v


~ command flips the case of the selected text

selection can also be combined with movement commands like w for word or e for end of word

numbers in vim
4j to go down 4 lines

in visual mode 3e to select 3 ends of words

and ofcourse these can be combined with deleting commands so we can do 7dw to delete 7 words

one final category of keys in vim is called modifiers, so far we have movements, edits, counts and finally we have modifiers

2 useful modifiers are a and i called around and inside

for example if we have a link in markdown like so [text to display](link)
and our cursor is somewhere inside the square brackets, and we want to change the text inside the square brackets we can do

way 1:
b - to go backward
3dw - delete 3 words
i - insert mode

way 2:
b - backward
3cw - change 3 words (automatically puts you in insert mode)

way 3:
ci[ - change inside [ (will remove everything inside [] and put you in insert mode

btw if your cursor is on one of the paranthesis (opening or closing you can press \% to jump between paranthesis

if we do da( it will delete everything inside the paranthesis including the paranthesis
